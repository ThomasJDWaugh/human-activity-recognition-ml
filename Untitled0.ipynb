```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv("/metadata.csv")
test = pd.read_csv("/metadata_test.csv")
```

```python
X_train = train.drop(columns=["user_snippet", "activity"])
y_train = train["activity"]
X_test = test.drop(columns=["user_snippet", "activity"])
```

```python
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train.drop('activity', axis=1),
                                                    train['activity'],
                                                     random_state=123)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)
y_train_encoded = label_encoder.fit_transform(y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=20,min_samples_split=15, min_samples_leaf=10, random_state=42)
clf.fit(X_train_imputed,y_train_encoded)
```

```python
y_pred = clf.predict(X_test_imputed)
```

```python
predicted_df = test.copy()
```

```python
predicted_df["activity"] = label_encoder.inverse_transform(y_pred)
```

```python
import os
predicted_file_path = "/Users/aritwaug/OneDrive - Liverpool John Moores University/Documents/7021datsci-challenge-2025/metadata_rfpredict.csv"
directory = os.path.dirname(predicted_file_path)

# Create the directory if it doesn't exist
if not os.path.exists(directory):
    os.makedirs(directory)

# Now save the DataFrame
predicted_df.to_csv(predicted_file_path, index=False)
print(f"Predicted file saved as: {predicted_file_path}")
```

```python
from sklearn.model_selection import learning_curve
import numpy as np
import matplotlib.pyplot as plt
# Generate learning curve
train_sizes, train_scores, test_scores = learning_curve(
    clf, X_train_imputed, y_train_encoded, cv=5, scoring="accuracy", train_sizes=np.linspace(0.1, 1.0, 10)
)

# Compute mean and std deviation
train_mean = np.mean(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)

# Plot learning curve
plt.plot(train_sizes, train_mean, label="Training Score", color="blue")
plt.plot(train_sizes, test_mean, label="Validation Score", color="red")
plt.fill_between(train_sizes, train_mean - np.std(train_scores, axis=1),
                 train_mean + np.std(train_scores, axis=1), color="blue", alpha=0.2)
plt.fill_between(train_sizes, test_mean - np.std(test_scores, axis=1),
                 test_mean + np.std(test_scores, axis=1), color="red", alpha=0.2)
plt.xlabel("Training Set Size")
plt.ylabel("Accuracy")
plt.title("Learning Curve")
plt.legend()
plt.show()
```
